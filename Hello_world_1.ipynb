```python
import tkinter as tk
from tkinter import ttk

win = tk.Tk()
win.title("python GUI")

l = ttk.Label(win, text="A Label")
l.grid(column=0, row=0)
def clickMe():
    action.configure(text="Hello" +' '+name.get() + ' '+ nc.get())
ttk.Label(win, text="Enter a name:").grid(column=0, row=0)

name = tk.StringVar()
nameE = ttk.Entry(win, width=12, textvariable=name)
nameE.grid(column=0, row=1)


action = ttk.Button(win, text='Click Me!', command=clickMe)
action.grid(column=2, row=1)

ttk.Label(win, text='Choose a number:').grid(column=1, row=0)
E = tk.StringVar()
nc = ttk.Combobox(win, width=12, textvariable=E, state='readonly')
nc['values'] = (1,2,4,42, 100)
nc.grid(column=1, row=1)
nc.current(0)

c1 = tk.IntVar()
ck1 = tk.Checkbutton(win, text='Disabled', variable= c1, state= 'disabled')
ck1.select()
ck1.grid(column=0, row=4, sticky=tk.W)

c2 = tk.IntVar()
ck2 = tk.Checkbutton(win, text='Enabled', variable= c2)
ck2.select()
ck2.grid(column=1, row=4, sticky=tk.W)

c3 = tk.IntVar()
ck3 = tk.Checkbutton(win, text='UnChecked', variable= c3)
ck3.select()
ck3.grid(column=2, row=4, sticky=tk.W)

clr1 = 'Blue'
clr2 = 'Gold'
clr3 = 'Red'

def color():
    r = c1.get()
    
    if r ==1: win.configure(background=clr1)
    elif r ==2: win.configure(background=clr2)
    elif r ==3: win.configure(background=clr3)    

c1 = tk.IntVar()
ck1 = tk.Radiobutton(win, text='Disabled', variable= c1, value=1, command=color)
ck1.grid(column=0, row=5, sticky=tk.W, columnspan=3)


ck2 = tk.Radiobutton(win, text='Enabled', variable= c1, value=2, command=color)
ck2.grid(column=1, row=5, sticky=tk.W, columnspan=3)

ck3 = tk.Radiobutton(win, text='UnChecked', variable= c1, value=3, command=color)
ck3.grid(column=2, row=5, sticky=tk.W, columnspan=3)

nameE.focus()
win.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import scrolledtext
win = tk.Tk()
win.title("python GUI")

l = ttk.Label(win, text="A Label")
l.grid(column=0, row=0)
def clickMe():
    action.configure(text="Hello" +' '+name.get() + ' '+ nc.get())
ttk.Label(win, text="Enter a name:").grid(column=0, row=0)

name = tk.StringVar()
nameE = ttk.Entry(win, width=12, textvariable=name)
nameE.grid(column=0, row=1)


action = ttk.Button(win, text='Click Me!', command=clickMe)
action.grid(column=2, row=1)

ttk.Label(win, text='Choose a number:').grid(column=1, row=0)
E = tk.StringVar()
nc = ttk.Combobox(win, width=12, textvariable=E, state='readonly')
nc['values'] = (1,2,4,42, 100)
nc.grid(column=1, row=1)
nc.current(0)


clr = ['Blue', 'Gold', 'Red']

def color():
    r = c1.get()
    
    if r ==0: win.configure(background=clr[0])
    elif r ==1: win.configure(background=clr[1])
    elif r ==2: win.configure(background=clr[2])    

c1 = tk.IntVar()
c1.set(99)

for c in range(3):
    crr = 'rad' + str(c)
    crr = tk.Radiobutton(win, text=clr[c], variable= c1, value=c, command=color)
    crr.grid(column=c, row=6, sticky=tk.W)



sW = 30
sH = 3

src = scrolledtext.ScrolledText(win, width=sW, height=sH, wrap=tk.WORD)
src.grid(column=0, columnspan=3)

nameE.focus()
win.mainloop()
```
